# 08 — Compare All Methods\n
\n
Aggregate metrics produced by each training notebook into comparison tables and plots.\n
\n
Outputs (planned):\n
- Detection: AP@[.50:.95], AP50, AP75 + loss curves\n
- Taxonomy: accuracy/F1 at family/genus/species; top-5; calibration + unknown threshold curves\n
- Few-shot curves (k vs metric)\n
- Federated vs centralized convergence curves

```python
from pathlib import Path
import json

import pandas as pd

ROOT = Path('..').resolve()
OUT = (ROOT / 'outputs').resolve()
print('outputs dir:', OUT)


def read_json(path: Path):
    return json.loads(path.read_text(encoding='utf-8'))


runs = []

# Transfer taxonomy
p = OUT / 'taxonomy_transfer' / 'val_metrics_last.json'
if p.exists():
    m = read_json(p)
    runs.append(
        {
            'method': 'transfer',
            'task': 'taxonomy',
            'val_species_acc_top1': m['species']['acc_top1'],
            'val_species_f1_macro': m['species']['f1_macro'],
        }
    )

# SSL taxonomy
p = OUT / 'taxonomy_ssl' / 'val_metrics_last.json'
if p.exists():
    m = read_json(p)
    runs.append(
        {
            'method': 'ssl_simclr',
            'task': 'taxonomy',
            'val_species_acc_top1': m['species']['acc_top1'],
            'val_species_f1_macro': m['species']['f1_macro'],
        }
    )

# Federated (FedAvg) taxonomy
p = OUT / 'federated_fedavg' / 'val_metrics_last.json'
if p.exists():
    m = read_json(p)
    runs.append(
        {
            'method': 'federated_fedavg',
            'task': 'taxonomy',
            'val_species_acc_top1': m['species']['acc_top1'],
            'val_species_f1_macro': m['species']['f1_macro'],
        }
    )

df = pd.DataFrame(runs)
df.sort_values(['task', 'method']) if len(df) else df
```
